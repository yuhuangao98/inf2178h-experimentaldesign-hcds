# Hypothesis Testing with Python

```python
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels
import numpy as np
from scipy import stats
import scipy.stats
import math
import random
import statistics
import scipy.stats as sp
```

### Now let's run a T-test

```python
print (sp.stats.ttest_ind(blackdf_male.dropna()['complainant_age_incident'], blackdf_male.dropna()['precinct']))
```

```python
print (sp.stats.ttest_ind(blackdf_male.dropna()['complainant_age_incident'], whitedf_male.dropna()['precinct']))
```

### (1) Starting with a 1-sample t test

```python
from scipy.stats import ttest_1samp
x =  blackdf_male.dropna()['complainant_age_incident']
tscore, pvalue = ttest_1samp(x, popmean=20)
print("t Statistic: ", tscore)  
print("P Value: ", pvalue)
```

### (2) Next: a 2-sample t test

```python
group1 = blackdf_male.dropna()['complainant_age_incident']
group2 = whitedf_male.dropna()['complainant_age_incident']
```

```python
#find variance for each group
print(np.var(group1), np.var(group2))
```

```python
#perform two sample t-test with equal variances
sp.ttest_ind(a=group1, b=group2, equal_var=True)
```

### Next let's intepret our results

The two hypotheses for this particular two sample t-test are as follows:

H0: µ1 = µ2 (the two population means are equal)

HA: µ1 ≠µ2 (the two population means are not equal)

Because the p-value of our test (4.25e-43) is greater than alpha = 0.05, we fail to reject the null hypothesis of the test. We do not have sufficient evidence to say that the mean complainant age between the two populations is different.

### (3) Paired Sample t-test

In order for the paired sample t-test results to be trusted, the following assumptions need to be met:

The dependent variable (DV) must be continuous which is measured on an interval or ratio scale
The observations are independent
The DV should be approximately normally distributed
The paired sample t-test is robust to this violation. If there is a violation of normality, as long as it’s not in a major violation the test results can be considered valid

```python
sp.ttest_rel(group1[:20], group2[:20])
```

#### Interpretation

A paired sample t-test was used to analyze the age of the complainants between the populations of black and white males to test if there was statistical significance between the age and number of allegations made by each population group. In comparing the two population groups, there was not a statistically significant relationship (p=0.255)

### (4) Welch's Sample t-test
Welch’s t-test is a nonparametric univariate test that tests for a significant difference between the mean of two unrelated groups. It is an alternative to the independent t-test when there is a violation in the assumption of equality of variances.

The hypothesis being tested is:

Null hypothesis (H0): u1 = u2, which translates to the mean of sample 1 is equal to the mean of sample 2
Alternative hypothesis (HA): u1 ≠ u2, which translates to the mean of sample 1 is not equal to the mean of sample 2
If the p-value is less than what is tested at, most commonly 0.05, one can reject the null hypothesis.

```python
stats.ttest_ind(group1, group2, equal_var = False)
```

#### Interpretation
The p-value is significant, therefore one can reject the null hypothesis in support of the alternative.
